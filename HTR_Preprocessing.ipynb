```python
# 2. Import libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# 3. Set image path (use the filename stored in your Colab directory)
img_path = 'test_image_unprocessed.png'  # Make sure this file exists in /content/

# 4. Load as grayscale
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# 5. Contrast enhancement with CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
enhanced = clahe.apply(img)



# 6. Visualize original, enhanced, and Otsu outputs
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("CLAHE Enhanced")
plt.imshow(enhanced, cmap='gray')
plt.axis('off')

# 8. Save result for OCR/HTR
cv2.imwrite('clahe_enhanced.png', enhanced)

print("Saved clahe_enhanced image as 'clahe_enhanced.png'")
```
